```
import ee
```

```
ee.Authenticate();
ee.Initialize();
```

```
# extra variables
tbudget = 1200 * 10**8 
tpowered = 100000
temissions = 22000 * 50

cbudget = 1.2 * 10**9
cpowered = 250000
cemissions = 2.5 * 10**6 * 50

sbudget = 10**10
spowered = 150000
semissions = 1500000 * 50

pbudget = 140 * 10**7
ppowered = 0
pemissions = -100 * 10**6
```

```
import random
import math
from IPython.display import Image, display

surfaceTemp = 0
budget = 0
budgetgoal = 5 * 10**12
powerGoal = 1 * 10**9
powered = 0
carbonemissions = 0
gtc = 0
yay = True



#funkshuns
def Intro():
  print("Welcome to this informative, totally not shitty climate change game! :D")
  print("You will choose what to build to power the houses you need to, and")
  print("then we will you crush you with the existential weight of your actions by showing you how you've failed the planet")
  print("You have a budget of 5 trillion, and you need to power 1 billion households! Manage your money but still power the houses accordingly")
  print("Nice! Let's begin!")

def WindTurbino(din):
  global budget
  global budgetgoal
  global powered
  global carbonemissions
  global yay
  newbudget = budget + tbudget * din
  maxamnt = int((budgetgoal - budget)/tbudget)
  if newbudget > budgetgoal:
    print("Oops, you can't do that! It exceeds your budget!")
    print("The maximum amount you can purchase here is: " + str(maxamnt)  )
    yay = False
  else:
    budget = newbudget
    carbonemissions += temissions * din
    powered += tpowered * din
  return;


def CoalMinero(din):
  global budget
  global budgetgoal
  global powered
  global carbonemissions
  global yay
  newbudget = budget + cbudget * din
  maxamnt = int((budgetgoal - budget)/cbudget)
  if newbudget > budgetgoal:
    print("Oops, you can't do that! It exceeds your budget!")
    print("The maximum amount you can purchase here is: " + str(maxamnt)  )
    yay = False
  else:
    budget = newbudget
    carbonemissions += cemissions * din
    powered += cpowered * din
  return;


def Policerino(din):
  global budget
  global budgetgoal
  global powered
  global carbonemissions
  global yay
  newbudget = budget + pbudget * din
  maxamnt = int((budgetgoal - budget)/pbudget)
  if newbudget > budgetgoal:
    print("Oops, you can't do that! It exceeds your budget!")
    print("The maximum amount you can purchase here is: " + str(maxamnt))
    yay = False
  else:
    budget = newbudget
    carbonemissions += pemissions * din
    powered += ppowered * din


def SolarPanelerino(din):
  global budget
  global budgetgoal
  global powered
  global carbonemissions
  global yay
  newbudget = budget + sbudget * din
  maxamnt = int((budgetgoal - budget)/sbudget)
  if newbudget > budgetgoal:
    print("Oops, you can't do that! It exceeds your budget!")
    print("The maximum amount you can purchase here is: " + str(maxamnt))
    yay = False
  else:
    budget = newbudget
    carbonemissions += semissions * din
    powered += spowered * din

def choice(userinput, din):
  if userinput == 'a':
    WindTurbino(din)
    return True
  elif userinput == 'b':
    CoalMinero(din)
    return True
  elif userinput == 'c':
    Policerino(din)
    return True
  elif userinput == 'd':
    SolarPanelerino(din)
    return True
  else:
    print("Invalid input. Try again")
    return False



# DISPLAY EARTH ENGINE

def cumco2ToYear(v_cumco2):
    return math.floor((65/500) * v_cumco2 + 1950)

def TempIncrFromCO2(v_cumco2):
    temp = (1.1/500) * v_cumco2
    return str(round(temp, 2))

def DisplayMap(cco2):

    vizparams = {
        'dimensions': 500,
        'min': 230.0,
        'max': 308.0,
        'palette': ['800080', '0000ab', '0000ff', '008000', '19ff2b', 'a8f7ff', 'ffff00', 'd6d600', 'ffa500', 'ff6b01', 'ff0000']
    }

    beginYearNum = cumco2ToYear(cco2)
    beginDateTemp = str(beginYearNum) + '-01-01'
    endDateTemp = str(beginYearNum) + '-01-02'


    dataset = ee.ImageCollection('NCEP_RE/surface_temp').filter(ee.Filter.date(beginDateTemp, endDateTemp))
    landSurfaceTemperature = dataset.select('air')
    img = landSurfaceTemperature.first()
    display(Image(url = img.getThumbUrl(vizparams)))
    print('Your chosen data correspond to the real year of ' + str(beginYearNum) + '! (Data: NCEP/NCAR Reanalysis Data, Surface Temperature)')
    print('So far, you\'ve increased average surface temperature since 1950 by ' + TempIncrFromCO2(cco2) + ' degrees Celsius!')



#gamestate shit
Gamerunning = True
running = True
Intro()
while Gamerunning: 
  surfaceTemp = 0
  budget = 0
  budgetgoal = 5 * 10**12
  powerGoal = 1 * 10**9
  powered = 0
  carbonemissions = 0
  gtc = 0 
  while running:
    buychoice = input("\nMake a choice: \n a) Build a wind turbine (Cost: 12 Billion USD, Houses powered: 100 Thousand) \n b) Build a coal mine (Cost: 1.2 Billion USD, Houses powered: 250 Thousand) \n c) Remove 100 million tons of CO2 from atmosphere (Cost: 1.4 Billion USD) \n d) Build solar farm (Cost: 10 Billion USD, Houses powered: 150 Thousand) \n")
    numberchoice = int(input("How many times will you do that? \n"))
    if choice(buychoice, numberchoice) == True and yay == True:
      remainingbudget = (budgetgoal - budget)
      print("You now have a remaining budget of $" + str(remainingbudget))
      print("You have used " + str(100-(100*remainingbudget / budgetgoal)) + "% of your budget.")
      print("You have powered " + str(powered) + " of " + str(powerGoal) + " houses")
      print("You have powered " + str(100*powered/powerGoal) + "% of all the houses.")
      gtc = carbonemissions / 1000000000
      if(gtc > 500):
        print("You have exceeded emissions up to present day.")
        running = False;
        break;
      
      DisplayMap(gtc)
      if remainingbudget < tbudget and remainingbudget < cbudget and remainingbudget < sbudget and remainingbudget < pbudget and powered < powerGoal:
        print ("You were unable to power all the houses. Now, let's take a look at the impact on the Earth!")
        running = False
      elif powered >= powerGoal : 
        print("Congrats! You were able to power all the houses. Now, let's take a look at the impact on the Earth!")
        running = False
    else:
      yay = True
  yaynay = input("Play again? ")
  if yaynay == "yes":
    running = True
  else:
    running = False
    Gamerunning = False


```
